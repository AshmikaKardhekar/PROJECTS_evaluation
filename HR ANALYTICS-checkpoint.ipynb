Our data seems to have many outliers as well as skewness. We will first deal with the skewness and then remove the outliers from the data. Then we will check the correlation between them and then decide which columns to keep/remove.

Most of the data is skewed to the left,so let's deal with it and have a normally distributed data.

Also the data is having outliers but not on a large scale  but the outlier distance is more so it tells us that there is more deflection present in the data than more number of outliers.

Let's deal with skewness & outliers!

```python
HR.skew()
```

We will remove the skewness of those columns whose values are more than 0.5.

```python
for s in HR.columns:
    if HR[s].dtype == 'int64':
        if HR[s].skew() > 0.5:
            HR[s] = np.log1p(HR[s])
        if HR[s].skew() < -0.5:
            HR[s] = np.square(HR[s])
HR.skew()
```

```python
HR.hist(grid=True, layout=(5,7), figsize=(20,20), bins=10, color='Orange')
plt.show()
```

```python
HR.plot(kind='box', subplots=True, layout=(5,7), figsize=(20,20), color='Orange')
plt.show()
```

We can see that most of the skewness is removed from our data as well as the outliers are also reduced because of the more normal distribution of the data. So, there is no such need of removing any outliers from out data.

# TRAINING THE MODEL

# Classification :-

Since, we have to predict the attrition, ie.whether the person will stay in the company or not. And our target variable is of categorical type. We'll use a classification model for making our predictions of Yes/No.

```python
d = ['Attrition', 'JobLevel', 'PerformanceRating', 'YearsInCurrentRole', 'YearsWithCurrManager']
X = HR.drop(d,axis=1)
X.head()
```

But first we'll have to use One-Hot-Encoder for giving 0's & 1's to our data with the object; we can't label encode them as they do not follow any ordinality.

```python
X = pd.get_dummies(X)
X.head()
```

```python
X.info()
```

After this we'll do the standardization of data because our data is not on the same scale.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
```

```python
X.head()
```

```python
y = HR['Attrition']
y.head()
```

```python
X.shape, y.shape
```

Splitting our data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=None)

#Importing Base Algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

#Importing Error Metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
model = [LogisticRegression(), KNeighborsClassifier(), GaussianNB(), SVC(), DecisionTreeClassifier()]

for m in model:
    m.fit(X_train, y_train)
    print('Score of', m, 'is:', m.score(X_train, y_train))
    pred = m.predict(X_test)
    print('AS:',accuracy_score(pred,y_test))
    print('CM:',confusion_matrix(pred,y_test))
    print('\n')
    print('.......................................................................................................')
    print('\n')
```

The model score of DecisionTree is highest followed by SVC and LogisticRegression.

But when we check the accuracy scores of all three of them we have a confusion between SVC & LogisticRegression.

We will now have to tune the parameters of LogisticRegression & SVC  and then choose which base algorithm works best after tunning.

Using GridSearchCV for parameter tunning of Base Algorithms.

1. LogisticRegression()

```python
from sklearn.model_selection import GridSearchCV
LR = LogisticRegression()
param_grid_LR = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'penalty': ['l1', 'l2'],
    'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']
}
GCV_LR = GridSearchCV(estimator=LR,param_grid=param_grid_LR,cv=10)
GCV_LR.fit(X_train,y_train)
```

```python
print('\n Best parameters = ')
print(GCV_LR.best_params_)
```

```python
print('Best Scores = ',GCV_LR.best_score_)
```

```python
#Feeding the best parameters to the Base algorithm.
LR_best = LogisticRegression(C=0.1, penalty='l2', solver='newton-cg')
LR_best.fit(X_train,y_train)
pred_LR = LR_best.predict(X_test)
print('Score of', LR_best, 'is:', LR_best.score(X_train, y_train))
print('AS:',accuracy_score(pred_LR,y_test))
print('CM:',confusion_matrix(pred_LR,y_test))
aucroc_scr_LR = roc_auc_score(y_test,pred_LR)
print('auc roc score is', aucroc_scr_LR)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_LR = cross_val_score(LR, X_train, y_train, cv=5, scoring='accuracy')
print('Score:',score_LR)
print('Mean Score:',score_LR.mean())
print('Standard deviation:',score_LR.std())
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_LR = cross_val_score(LR, X_train, y_train, cv=5, scoring='roc_auc')
print('Score:',score_LR)
print('Mean Score:',score_LR.mean())
print('Standard deviation:',score_LR.std())
```

2. SVC

```python
from sklearn.model_selection import GridSearchCV
svc = SVC()
param_grid_svc = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'kernel':['linear', 'poly', 'rbf'], 
}
GCV_svc = GridSearchCV(estimator=svc,param_grid=param_grid_svc,cv=5)
GCV_svc.fit(X_train,y_train)
```

```python
print('\n Best parameters = ')
print(GCV_svc.best_params_)
```

```python
print('Best Scores = ',GCV_svc.best_score_)
```

```python
#Feeding the best parameters to the Base algorithm.
svc_best = SVC(C=0.1, kernel='linear', probability=True)
svc_best.fit(X_train,y_train)
pred_svc = svc_best.predict(X_test)
print('Score of', svc_best, 'is:', svc_best.score(X_train, y_train))
print('AS:',accuracy_score(pred_svc,y_test))
print('CM:',confusion_matrix(pred_svc,y_test))
aucroc_scr_svc = roc_auc_score(y_test,pred_svc)
print('auc roc score is', aucroc_scr_svc)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_svc = cross_val_score(svc, X_train, y_train, cv=5, scoring='accuracy')
print('Score:',score_svc)
print('Mean Score:',score_svc.mean())
print('Standard deviation:',score_svc.std())
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_svc = cross_val_score(svc, X_train, y_train, cv=5, scoring='roc_auc')
print('Score:',score_svc)
print('Mean Score:',score_svc.mean())
print('Standard deviation:',score_svc.std())
```

We see that all the scores of both the algorithms are almost same.

So, from the above observations we will select SVC as our final base model and hence move forward to the ensembling techniques.

Let's use Ensembling Techniques!

We'll use any one of the boosting techniques and then tune it, once done we will then fit our final model.

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier(base_estimator=svc_best)
param_grid_ada = {
    'n_estimators':[50,100,150,200,250,300],
    'algorithm':['SAMME', 'SAMME.R']
}
GCV_ada = GridSearchCV(estimator=ada,param_grid=param_grid_ada,cv=5)
GCV_ada.fit(X_train,y_train)
```

```python
print('\n Best parameters = ')
print(GCV_ada.best_params_)
```

```python
print('Best Scores = ',GCV_ada.best_score_)
```

```python
#Feeding the best parameters.
ada_best = AdaBoostClassifier(base_estimator=svc_best,n_estimators=50, algorithm='SAMME')
ada_best.fit(X_train,y_train)
pred_ada = ada_best.predict(X_test)
print('Score of', ada_best, 'is:', ada_best.score(X_train, y_train))
print('AS:',accuracy_score(pred_ada,y_test))
print('CM:',confusion_matrix(pred_ada,y_test))
aucroc_scr_ada = roc_auc_score(y_test,pred_ada)
print('auc roc score is', aucroc_scr_ada)
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_ada = cross_val_score(ada_best, X_train, y_train, cv=5, scoring='accuracy')
print('Score:',score_ada)
print('Mean Score:',score_ada.mean())
print('Standard deviation:',score_ada.std())
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_ada = cross_val_score(ada_best, X_train, y_train, cv=5, scoring='roc_auc')
print('Score:',score_ada)
print('Mean Score:',score_ada.mean())
print('Standard deviation:',score_ada.std())
```

After tunning the base estimator and using ensembling techniques our model is not working well since all the scores have decreased.
Though the accuracy quite good the auc-roc curve isn't given a good fit which means our model isn't predicting much true positive values

So, we will use SVC after tunning as the final model and fit 100% data into it.
