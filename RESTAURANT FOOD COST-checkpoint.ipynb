# TRAINING THE MODEL

# Regression :-

Since we have to predict the cost of the dishes served by the restaurant, our problem of regression type.

```python
X = cost.drop('COST', axis=1)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
X = pd.get_dummies(X)
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
```

```python
X.head()
```

```python
y = cost['COST']
y.head()
```

```python
X.shape , y.shape
```

```python
from sklearn.model_selection import train_test_split

#Importing Base Algorithms
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

#Importing boosting algorithms
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from xgboost.sklearn import XGBRegressor
from lightgbm import LGBMRegressor

#Importing Error Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = None, test_size = 0.30)
model_r = [LinearRegression(), DecisionTreeRegressor(), KNeighborsRegressor(), SVR(), Lasso(), Ridge(), ElasticNet(), RandomForestRegressor(), AdaBoostRegressor(), GradientBoostingRegressor(), XGBRegressor(verbosity=0), LGBMRegressor(silent=True)]

for m in model_r:
    m.fit(X_train, y_train)
    print('Score of', m, 'is:', m.score(X_train, y_train))
    pred_r = m.predict(X_test)
    
    print('error:')
    print('Mean absolute error:',mean_absolute_error(y_test,pred_r))
    print('Mean squared error:',mean_squared_error(y_test,pred_r))
    print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred_r)))
    print('r2_score:',r2_score(y_test,pred_r))
    print('\n')
    print('.......................................................................................................')
    print('\n')
```

We have various models who have almost similar r2 scores, let's run a loop to get the max r2 score which will give us the max r2 score to the corresponding best random state of that particular model. 

```python
model_r = [LinearRegression(), DecisionTreeRegressor(), KNeighborsRegressor(), SVR(), Lasso(), Ridge(), ElasticNet(), RandomForestRegressor(), AdaBoostRegressor(), GradientBoostingRegressor(), XGBRegressor(verbosity=0), LGBMRegressor(silent=True)]

max_r2score = 0
for r_state in range(0,100):
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = r_state, test_size = 0.30)
        for m in model_r:
            m.fit(X_train, y_train)
            pred_r = m.predict(X_test)
            r2score = r2_score(y_test,pred_r)
            print('r2score corresponding to the random state', r_state, 'is', r2score)
            if r2score > max_r2score:
                max_r2score = r2score
                final_r_state = r_state
                final_model = m
print()
print()
print('max r2score corresponding to random state', final_r_state, 'is', max_r2score, 'and the final model is', final_model)
            
```

We have our max r2 score of 77.47% given by LGBMRegressor() at a random state=5.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 5, test_size = 0.30)
LGBM = LGBMRegressor(silent=True)
LGBM.fit(X_train,y_train)
pred = LGBM.predict(X_test)
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))
print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred)))
print('r2_score:',r2_score(y_test,pred))
```

Let's apply gird search cv to check the performance of the model and tune the parameters for better performance, as it will give us the best parameters which will ensure that our model does not overfit or underfit.

```python
from sklearn.model_selection import GridSearchCV
param_grid = {"max_depth": [2,4,6,8,10],
              "learning_rate" : [0.01,0.05,0.1],
              "num_leaves": [100,200,300,400,500],
              "n_estimators": [50,100,150,200]
             }
GCV = GridSearchCV(LGBM,param_grid=param_grid, cv = 5, scoring="r2")
GCV.fit(X_train,y_train)
```

```python
print('\n Best parameters = ')
print(GCV.best_params_)
```

```python
print('Best Scores = ',GCV.best_score_)
```

```python
#Feeding the best parameters to the LGBM
LGBM_best = LGBMRegressor(learning_rate=0.01, max_depth=4, n_estimators=200, num_leaves=100)
LGBM_best.fit(X_train,y_train)
pred_LGBM = LGBM_best.predict(X_test)
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred_LGBM))
print('Mean squared error:',mean_squared_error(y_test,pred_LGBM))
print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred_LGBM)))
print('r2_score:',r2_score(y_test,pred_LGBM))
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score = cross_val_score(LGBM_best, X_train, y_train, cv=5, scoring='r2')
print('Score:',score)
print('Mean Score:',score.mean())
print('Standard deviation:',score.std())
```
