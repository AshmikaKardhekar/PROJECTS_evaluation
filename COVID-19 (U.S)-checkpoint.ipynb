# TRAINING THE MODEL

# Regression :-

Since our target is to find out the hospitalization rate and it has continuous data in it.

```python
X = C19.drop('Hospitalization_Rate',axis=1)
X.head()
```

Since, our data has different units we will have to bring them on the same scale and hence we'll do standardization of data.

Also we'll have to do one hot encoding because our data is not ordinal and we don't want our model to give weights according to the numbers mentioned in individual columns.

```python
C19.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
X = pd.get_dummies(X, columns = ['Province_State'])
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
```

```python
X.head()
```

```python
y = C19['Hospitalization_Rate']
y.head()
```

```python
X.shape, y.shape
```

Applying PCA.

```python
#Principal Component Analysis
from sklearn.decomposition import PCA
pca_1 = PCA(n_components = 4)
cases = pd.DataFrame(pca_1.fit_transform(X[['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'People_Tested', 'People_Hospitalized']]), columns=['pca_1','pca_2','pca_3','pca_4'])
pca_2 = PCA(n_components = 1)
date = pd.DataFrame(pca_2.fit_transform(X[['Month', 'Week']]), columns=['pca_5'])
X = pd.concat((X, cases, date), axis=1)
X.drop(['Confirmed', 'Deaths', 'Recovered', 'Active', 'Incident_Rate', 'People_Tested', 'People_Hospitalized','Month', 'Week'], axis=1, inplace=True)
X.head()
```

Splitting our data.

```python
from sklearn.model_selection import train_test_split

#Importing Base Algorithms
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

#Importing Error Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
model_r = [LinearRegression(), DecisionTreeRegressor(), KNeighborsRegressor(), SVR(), Lasso(), Ridge(), ElasticNet()]
max_r2score = 0
for r_state in range(0,100):
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = r_state, test_size = 0.30)
        for m in model_r:
            m.fit(X_train, y_train)
            pred_r = m.predict(X_test)
            r2score = r2_score(y_test,pred_r)
            print('r2score corresponding to the random state', r_state, 'is', r2score)
            if r2score > max_r2score:
                max_r2score = r2score
                final_r_state = r_state
                final_model = m
print()
print()
print('max r2score corresponding to random state', final_r_state, 'is', max_r2score, 'and the final model is', final_model)
            
```

So we have our best r2_score of 0.98 with it's random state as 32 we will now try and tune the parameters of our base model ie.KNeighborsRegressor.

random_state as the name suggests, is used for initializing the internal random number generator, which will decide the splitting of data into train and test indices in your case. If we choose the best random set and feed it to our model the assumptions and predictions made by our model whould be the best.

Let's tune the parameters!

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 32, test_size = 0.30)
knr = KNeighborsRegressor()
knr.fit(X_train,y_train)
pred = knr.predict(X_test)
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))
print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred)))
print('r2_score:',r2_score(y_test,pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_neighbors':[2,4,6,8,10],
              'weights':['uniform', 'distance'],
              'algorithm':['auto', 'ball_tree','kd_tree','brute']}
GCV_knr = GridSearchCV(estimator=knr,param_grid=param_grid,cv=5)
GCV_knr.fit(X_train,y_train)
```

```python
print('\n Best parameters = ')
print(GCV_knr.best_params_)
```

```python
print('Best Scores = ',GCV_knr.best_score_)
```

```python
#Feeding the best parameters to the Base algorithm.
knr_best = KNeighborsRegressor(algorithm = 'brute', n_neighbors = 2, weights = 'distance')
knr_best.fit(X_train,y_train)
pred_knr = knr_best.predict(X_test)
print('Score of', knr_best, 'is:', knr_best.score(X_train, y_train))
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred_knr))
print('Mean squared error:',mean_squared_error(y_test,pred_knr))
print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred_knr)))
print('r2_score:',r2_score(y_test,pred_knr))
```
