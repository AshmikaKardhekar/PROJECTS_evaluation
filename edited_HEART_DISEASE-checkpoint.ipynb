# TRAINING THE MODEL

# Classification :-

The "num" columns refers to the presence of heart disease in the patient.
Since, we have to predict these numbers in terms of 'distinguishing the presence of disease (values 1,2,3,4) from absence of the disease (value 0).'; we will be using a classification model to make the predictions.

```python
X = HD.drop('num',axis=1)
X.head()
```

Since, our data has different units we will have to bring them on the same scale and hence we'll do standardization of data.

Also we'll have to do one hot encoding because our data is not ordinal and we don't want our model to give weights according to the numbers mentioned in individual columns.

```python
from sklearn.preprocessing import StandardScaler
X = pd.get_dummies(X, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'])
sc = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
X[columns_to_scale] = sc.fit_transform(X[columns_to_scale])
```

```python
X.head()
```

```python
y = HD['num']
y.head()
```

```python
X.shape, y.shape
```

Splitting our data.

```python
from sklearn.model_selection import train_test_split

#Importing Base Algorithms
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

#Importing Error Metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
model = [KNeighborsClassifier(), GaussianNB(), SVC(), DecisionTreeClassifier()]
max_ac_score = 0
for r_state in range(0,100):
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = r_state, test_size = 0.20)
        for m in model:
            m.fit(X_train, y_train)
            pred = m.predict(X_test)
            ac_score = accuracy_score(y_test,pred)
            print('Accuracy_score corresponding to the random state', r_state, 'is', ac_score)
            if ac_score > max_ac_score:
                max_ac_score = ac_score
                final_r_state = r_state
                final_model = m
print()
print()
print('max ac_score corresponding to random state', final_r_state, 'is', max_ac_score, 'and the final model is', final_model)
```

So we have our best accuracy of 77.04% with it's random state as 41, we will now try and tune the parameters of our base model ie.KNeighborsClassifier.

Let's try and tune the number of neighbors.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 41, test_size = 0.20)
KNN_scores = []
for k in range(1,50):
    KNN_best = KNeighborsClassifier(n_neighbors = k)
    KNN_best.fit(X_train, y_train)
    KNN_scores.append(KNN_best.score(X_test, y_test))
```

```python
plt.figure(figsize=(20,10))
plt.plot([k for k in range(1, 50)], KNN_scores, color = 'red')
plt.xticks([i for i in range(1, 50)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

So at number of neighbors = 8 we are getting the highest accuracy.

```python
#Feeding the best parameters to the Base algorithm.
KNN_best = KNeighborsClassifier(n_neighbors = 8)
KNN_best.fit(X_train,y_train)
pred_KNN = KNN_best.predict(X_test)
print('Score of', KNN_best, 'is:', KNN_best.score(X_train, y_train))
print('AS:',accuracy_score(pred_KNN,y_test))
print('CM:',confusion_matrix(pred_KNN,y_test))
```

Our accuracy increased from 77.04% to 78.68% which is quiet helpful for our models learning.

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_KNN = cross_val_score(KNN_best, X_train, y_train, cv=10, scoring='accuracy')
print('Score:',score_KNN)
print('Mean Score:',score_KNN.mean())
print('Standard deviation:',score_KNN.std())
```

Here, the model's score was 63.22% after tunning it's parameter & now when we are checking using cross validation it is about 56.16%; which means our model is performing fairly well.

We can now say that our model is neither underfitting nor overfiiting.

Since our model is not performing well, we should try some other approach so that our model performs well, as we have very less information(rows) present in our data which is not giving much information to the model and hence the model's performance is hampered a bit.

Let's try and make some changes in the target variable ie.num.

We have very less difference in the number of people who are diagnosed with no heart diease and the total number of people who are diagnosed with the heart disease with type 1,2,3,4 combined together.

So, let's train our model in such a way that our model has to predict whether the person is having a heart disease or not.

For doing so we'll keep :- 
            
            0 = people with no heart disease
            (1,2,3,4) = 1 = people with heart disease
        
Now, we'll see how our model performs.

```python
HD['num'] = HD['num'].replace([2,3,4],1)
y_new = HD['num']
y_new.head()
```

```python
HD['num'].value_counts()
```

```python
model = [LogisticRegression(), KNeighborsClassifier(), GaussianNB(), SVC(), DecisionTreeClassifier()]
max_ac_score = 0
for r_state in range(0,100):
        X_train, X_test, y_new_train, y_new_test = train_test_split(X, y_new, random_state = r_state, test_size = 0.20)
        for m in model:
            m.fit(X_train, y_new_train)
            pred = m.predict(X_test)
            ac_score = accuracy_score(y_new_test,pred)
            print('Accuracy_score corresponding to the random state', r_state, 'is', ac_score)
            if ac_score > max_ac_score:
                max_ac_score = ac_score
                final_r_state = r_state
                final_model = m
print()
print()
print('max ac_score corresponding to random state', final_r_state, 'is', max_ac_score, 'and the final model is', final_model)
```

So we have our best accuracy of 95.08% with it's random state as 46, we will now try and tune the parameters of our base model ie.SVC.

```python
X_train, X_test, y_new_train, y_new_test = train_test_split(X, y_new, random_state = 46, test_size = 0.20)
from sklearn.model_selection import GridSearchCV
svc = SVC()
param_grid_svc = {
    'C': [0.001, 0.01, 0.1, 1, 10, 100],
    'kernel':['linear', 'poly', 'rbf'], 
}
GCV_svc = GridSearchCV(estimator=svc,param_grid=param_grid_svc,cv=5)
GCV_svc.fit(X_train,y_new_train)
```

```python
print('\n Best parameters = ')
print(GCV_svc.best_params_)
```

```python
print('Best Scores = ',GCV_svc.best_score_)
```

```python
#Feeding the best parameters to the Base algorithm.
svc_best = SVC(C=0.01, kernel='linear', probability=True)
svc_best.fit(X_train,y_new_train)
pred_svc = svc_best.predict(X_test)
print('Score of', svc_best, 'is:', svc_best.score(X_train, y_new_train))
print('AS:',accuracy_score(y_new_test,pred_svc))
print('CM:',confusion_matrix(y_new_test,pred_svc,))
aucroc_scr_svc = roc_auc_score(y_new_test,pred_svc)
print('auc roc score is', aucroc_scr_svc)
```

So, the accuracy changed from 95% to 91% which is quite considerable because by doing grid search cv we came to know the performance of the model which tells us that our model is not prone to underfitting or overfitting.
