We have a list of unique values now, but the variety is wide. We will now segregate each feature according to the most common occured value which will be the degree wise seperation but we will group the sub categories of that degree into a main degree as we already have most of the information available in the profile column. This way would be more convinient as it would not confuse the model with excessive data but help it learn the data more appropriately.

```python
#Let's check the maximum length of the total degrees of education first.
max_length = 0
for i in range(len(fees['Qualification'])):
    temp = fees['Qualification'].iloc[i].strip().upper().split(',')
    if len(temp) > max_length:
        max_length = len(temp)
print(max_length)
```

We do have some data which seems to be inappropriate, rather than keeping it let's remove those rows and only keep the appropriate data for training. Let's find out such inappropriate information present in our data.

```python
degrees = np.where(fees['Qualification'].str.contains('(BSc|MSc|BDS|MDS|BHMS|BAMS|MBBS|MD|MS|DNB|PhD|DDV|DDVL)')==False)
for d in degrees:
    print(fees['Qualification'].iloc[d])
```

```python
fees['Qualification'].iloc[d].value_counts()
```

We have very miscellenous columns present in our data and seperating them would create a chaos, rather than categorizing them into individual columns let's add them into a column which contains all the random miscellenous information.

```python
# Let's seperate the columns
q = ['BSc','BDS','BHMS','BAMS','DNB','MBBS','MSc','MDS','MD','MS','PhD','DDVL','DDV']
for i in range(len(q)):
    temp = q[i]
    qualification = []
    for j in range(len(fees['Qualification'])):
        split_data = fees['Qualification'][j].split(',')
        dummy = 0
        for n in split_data:
            if n.__contains__(temp):
                dummy = dummy + 1
                fees['Qualification'][j] = fees['Qualification'][j].replace(n,'')
        if dummy > 0:
            qualification.append(1) 
        else:
            qualification.append(0)
    fees[temp] = qualification        
```

Since we have seperated the major qualification categories we will have to seperate out the extra information or minor categories of education in the form of certifications or any other type available. We will then store them into a new column named Assorted.

```python
others = []
for a in range(len(fees['Qualification'])):
    dummy = len(fees['Qualification'][a])
    if dummy > 2:
        others.append(1)
    else:
        others.append(0)
fees['Assorted'] = others
fees.drop('Qualification', axis=1, inplace=True)
```

```python
fees.head(1)
```

Now we have the education in terms of their degrees well categorised into a common platform and some other random info saved in the "Assorted" column.

We have more than 50% missing values , let's impute those missing values by 'missing' as we can't replace the NANs with any other number.
Reason being the rating is exclusive of various factors and it follows ordinality and replacing the NANs with any other value may cause bais in the data which we don't want.

```python
fees['Rating'].fillna('Missing%', inplace=True)
fees['Rating']
```

Let's drop out the % symbol first and convert that column into an integer type.

```python
# Let's remove the % sign
fees['Rating'] = fees['Rating'].str.strip('%')
fees['Rating']
```

Let's create groups of ratings for better understanding of the data, it will also help the model to learn faster as we have a lot of variety present in our data. 

```python
for i in range(len(fees['Rating'])):
    if fees['Rating'].iloc[i] != 'Missing':
        rating = int(fees['Rating'].iloc[i])
        if rating in range(0,21):
            fees['Rating'].iloc[i] = '1-Star'
        if rating in range(21,41):
            fees['Rating'].iloc[i] = '2-Star'
        if rating in range(41,61):
            fees['Rating'].iloc[i] = '3-Star'
        if rating in range(61,81):
            fees['Rating'].iloc[i] = '4-Star'
        if rating in range(81,101):
            fees['Rating'].iloc[i] = '5-Star'
fees['Rating'].value_counts()
```

# TRAINING THE MODEL

# Regression :-

Since we have to predict the fees charged by the doctor, Fees becomes our target variable and hence this is a problem of regression.

```python
X = fees.drop('Fees', axis=1)
X.head()
```

```python
X.columns
```

```python
y = fees['Fees']
y.head()
```

```python
fees['Fees'].value_counts
```

```python
X.shape , y.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
for col in X.columns:
    if X[col].dtype=="object":
        X[col]=LabelEncoder().fit_transform(X[col])
print(X)
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split

#Importing Base Algorithms
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

#Importing boosting algorithms
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from xgboost.sklearn import XGBRegressor

#Importing Error Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = None, test_size = 0.30)
model_r = [LinearRegression(), DecisionTreeRegressor(), KNeighborsRegressor(), SVR(), Lasso(), Ridge(), ElasticNet(), RandomForestRegressor(), AdaBoostRegressor(), GradientBoostingRegressor(), XGBRegressor(verbosity=0)]

for m in model_r:
    m.fit(X_train, y_train)
    print('Score of', m, 'is:', m.score(X_train, y_train))
    pred_r = m.predict(X_test)
    
    print('error:')
    print('Mean absolute error:',mean_absolute_error(y_test,pred_r))
    print('Mean squared error:',mean_squared_error(y_test,pred_r))
    print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred_r)))
    print('r2_score:',r2_score(y_test,pred_r))
    print('\n')
    print('.......................................................................................................')
    print('\n')
```

We can see that the GradientBoostingRegressor is working the best out of all as it's r2 score is highest among all as well as the errors are low.

Let's choose the best random state for GradientBoostingRegressor as it is working the best in terms of r2 score.

But first let's see the difference in the r2 scores of 2 models which are performing very close.

```python
model_r = [GradientBoostingRegressor(),XGBRegressor(verbosity=0)]
max_r2score = 0
for r_state in range(0,100):
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = r_state, test_size = 0.30)
        for m in model_r:
            m.fit(X_train, y_train)
            pred_r = m.predict(X_test)
            r2score = r2_score(y_test,pred_r)
            print('r2score corresponding to the random state', r_state, 'is', r2score)
            if r2score > max_r2score:
                max_r2score = r2score
                final_r_state = r_state
                final_model = m
print()
print()
print('max r2score corresponding to random state', final_r_state, 'is', max_r2score, 'and the final model is', final_model)
            
```

max r2score corresponding to random state 44 is 0.2426291245795954 and the final model is GradientBoostingRegressor()

Let's use GradientBoostingRegressor() as our model.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 44, test_size = 0.30)
GB = GradientBoostingRegressor()
GB.fit(X_train,y_train)
pred = GB.predict(X_test)
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))
print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred)))
print('r2_score:',r2_score(y_test,pred))
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score = cross_val_score(GB, X_train, y_train, cv=5, scoring='r2')
print('Score:',score)
print('Mean Score:',score.mean())
print('Standard deviation:',score.std())
```

```python
import lightgbm as lgb
model_r = [lgb.LGBMRegressor(silent=True)]
max_r2score = 0
for r_state in range(0,100):
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = r_state, test_size = 0.30)
        for m in model_r:
            m.fit(X_train, y_train)
            pred_r = m.predict(X_test)
            r2score = r2_score(y_test,pred_r)
            print('r2score corresponding to the random state', r_state, 'is', r2score)
            if r2score > max_r2score:
                max_r2score = r2score
                final_r_state = r_state
                final_model = m
print()
print()
print('max r2score corresponding to random state', final_r_state, 'is', max_r2score, 'and the final model is', final_model)
            
```

So the r2 score of LGBM is greater than that of GB, we'll try tunning the parameter and see the results of LGBM.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 89, test_size = 0.30)
LGBM = lgb.LGBMRegressor(silent=True)
LGBM.fit(X_train,y_train)
pred = LGBM.predict(X_test)
print('error:')
print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))
print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred)))
print('r2_score:',r2_score(y_test,pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {"max_depth": [25,50, 75],
              "learning_rate" : [0.01,0.05,0.1],
              "num_leaves": [300,900,1200],
              "n_estimators": [200]
             }
GCV = GridSearchCV(LGBM,param_grid=param_grid, cv = 5, scoring="r2")
GCV.fit(X_train,y_train)
```

```python
#We'll have to segregate the categorical data as the given data is not categorised in a way a model can understand and the data will be more complicated to be handled when we have this much variation in it, so let's start seperating the data within the columns as we have multiple information in it.
#Let's check the maximum length of the total degrees of education first.
max_length = 0
for i in range(len(test['Qualification'])):
    temp = test['Qualification'].iloc[i].strip().upper().split(',')
    if len(temp) > max_length:
        max_length = len(temp)
print(max_length)
```

```python
#We do have some data which seems to be inappropriate, rather than keeping it let's remove those rows and only keep the appropriate data for training. Let's find out such inappropriate information present in our data.
degrees = np.where(test['Qualification'].str.contains('(BSc|MSc|BDS|MDS|BHMS|BAMS|MBBS|MD|MS|DNB|PhD|DDV|DDVL)')==False)
for d in degrees:
    print(test['Qualification'].iloc[d])
```

```python
# Let's seperate the columns
q = ['BSc','BDS','BHMS','BAMS','DNB','MBBS','MSc','MDS','MD','MS','PhD','DDVL','DDV']
for i in range(len(q)):
    temp = q[i]
    qualification = []
    for j in range(len(test['Qualification'])):
        split_data = test['Qualification'][j].split(',')
        dummy = 0
        for n in split_data:
            if n.__contains__(temp):
                dummy = dummy + 1
                test['Qualification'][j] = test['Qualification'][j].replace(n,'')
        if dummy > 0:
            qualification.append(1) 
        else:
            qualification.append(0)
    test[temp] = qualification        
```
