# TRAINING THE MODEL

# Classification :-

Since we have to find the class type which may be a star, galaxy or quasar our problem turns out to be a classification problem.

Since, our data has different units we will have to bring them on the same scale and hence we'll do standardization of data.

```python
#Split the data first
X = SS.drop('class',axis=1)
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns=X.columns)
```

```python
X.head()
```

```python
#Principal Component Analysis
from sklearn.decomposition import PCA
pca = PCA(n_components = 3)
bands = pd.DataFrame(pca.fit_transform(X[['u','g','r','i','z']]), columns=['pca_1','pca_2','pca_3'])
X = pd.concat((X,bands), axis=1)
X.drop(['u','g','r','i','z'], axis=1, inplace=True)
X.head()
```

```python
y = SS['class']
y.head()
```

Splitting our data.

```python
from sklearn.model_selection import train_test_split

#Importing Base Algorithms
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

#Importing Error Metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
model = [KNeighborsClassifier(), GaussianNB(), SVC(), DecisionTreeClassifier()]
max_ac_score = 0
for r_state in range(0,100):
        X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = r_state, test_size = 0.30)
        for m in model:
            m.fit(X_train, y_train)
            pred = m.predict(X_test)
            ac_score = accuracy_score(y_test,pred)
            print('Accuracy_score corresponding to the random state', r_state, 'is', ac_score)
            if ac_score > max_ac_score:
                max_ac_score = ac_score
                final_r_state = r_state
                final_model = m
print()
print()
print('max ac_score corresponding to random state', final_r_state, 'is', max_ac_score, 'and the final model is', final_model)
```

So we have our best accuracy of 99.36% with it's random state as 53 we will now try and tune the parameters of our base model ie.DecisionClassifier.

Let's tune the parameters!

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 53, test_size = 0.30)
DT = DecisionTreeClassifier()
DT.fit(X_train,y_train)
pred = DT.predict(X_test)
print('Score of', DT, 'is:', DT.score(X_train, y_train))
print('AS:',accuracy_score(y_test,pred))
print('CM:',confusion_matrix(y_test,pred))
print('CR:',classification_report(y_test,pred))
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_DT = cross_val_score(DT, X_train, y_train, cv=5, scoring='accuracy')
print('Score:',score_DT)
print('Mean Score:',score_DT.mean())
print('Standard deviation:',score_DT.std())
```

```python
pred_prob = DT.predict_proba(X_test)[:,1]
fpr,tpr,thresholds = roc_curve(y_test,pred_prob,pos_label=True)
plt.plot([0,1], [0,1], 'k--')
plt.plot(fpr, tpr)
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('AUC-ROC')
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
param_grid_DT = {'max_features': ['log2', 'sqrt','auto'], 
              'criterion': ['entropy', 'gini'],
              'max_depth': [2, 3, 5, 10, 50], 
              'min_samples_split': [2, 3, 50, 100],
              'min_samples_leaf': [1, 5, 8, 10]
             }
GCV_DT = GridSearchCV(estimator=DT,param_grid=param_grid_DT,cv=5)
GCV_DT.fit(X_train,y_train)
```

```python
print('\n Best parameters = ')
print(GCV_DT.best_params_)
```

```python
print('Best Scores = ',GCV_DT.best_score_)
```

```python
#Feeding the best parameters to the Base algorithm.
DT_best = DecisionTreeClassifier(criterion='gini', max_depth=10, max_features='sqrt', min_samples_leaf=1, min_samples_split=50)
DT_best.fit(X_train,y_train)
pred_DT = DT_best.predict(X_test)
print('Score of', DT_best, 'is:', DT_best.score(X_train, y_train))
print('AS:',accuracy_score(y_test,pred_DT))
print('CM:',confusion_matrix(y_test,pred_DT))
print('CR:',classification_report(y_test,pred_DT))
```

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_DT_best = cross_val_score(DT_best, X_train, y_train, cv=5, scoring='accuracy')
print('Score:',score_DT_best)
print('Mean Score:',score_DT_best.mean())
print('Standard deviation:',score_DT_best.std())
```
