# TRAINING THE MODEL

# Regression :-

PART 1 :-
    1. Predicting the average price of avocados.
    2. Average price is a continuous type of data.
    3. Hence, our problem is of Regression Type.

Daily sales of avocados is so uncertain, rather the effect on sales would be seasonal or due to the occurance of festivals in a week or so. It is therefore not necessarily important to keep the day column as it can be a redundant column in our dataset.

As per our data region & type do not follow any kind of ordinality and so if we apply label encoder to it whilst trainig our model; the model would show some biasing in the prediction depending on the weights give to the object data by label encoder.  

So, in order to avoid this situation we will use one hot encoding instead.

```python
a = ['AveragePrice','day']
X_R = avocado.drop(a,axis=1)
X_R.head()
```

```python
X_R['region'] = X_R['region'].astype('object')
X_R['type'] = X_R['type'].astype('object')
```

```python
X_R = pd.get_dummies(X_R)
X_R.head()
```

We'll use standard scaler for scaling our data as we have very different parameters throughout our data and so scaling them on one particular scale would increase the efficiency of the model's performance.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_R = pd.DataFrame(sc.fit_transform(X_R), columns=X_R.columns)
X_R.head()
```

```python
y_R = avocado['AveragePrice']
y_R.head()
```

Since the data is already within a specific range in case of the continuous type of variables, we don't need to apply standard scaler.

```python
X_R.shape, y_R.shape
```

Splitting our data.

```python
from sklearn.model_selection import train_test_split
X_R_train, X_R_test, y_R_train, y_R_test = train_test_split(X_R, y_R, test_size=0.20, random_state=None)
```

```python
X_R_train.shape, X_R_test.shape
```

```python
y_R_train.shape, y_R_test.shape
```

Model selection of Base Algorithms.

```python
#Importing Base Algorithms
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor

#Importing Error Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
model_r = [LinearRegression(), DecisionTreeRegressor(), KNeighborsRegressor(), SVR(), Lasso(), Ridge(), ElasticNet()]

for m in model_r:
    m.fit(X_R_train, y_R_train)
    print('Score of', m, 'is:', m.score(X_R_train, y_R_train))
    pred_r = m.predict(X_R_test)
    
    print('error:')
    print('Mean absolute error:',mean_absolute_error(y_R_test,pred_r))
    print('Mean squared error:',mean_squared_error(y_R_test,pred_r))
    print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_R_test,pred_r)))
    print('r2_score:',r2_score(y_R_test,pred_r))
    print('\n')
    print('.......................................................................................................')
    print('\n')
```

Although Decision tree has the highest score = 1 we can't be sure whether the algorithm is underfitting or it is a case of overfitting. So, we'll consider the algorithm having 2nd highest score.

We will choose KNearestNeighbors as our base algorithm because :-
    
    Mean absolute error: lowest among all
    
    Mean squared error: lowest among all
    
    Root Mean Squared Error: lowest among all
    
    r2_score: highest & closest to 1

Using GridSearchCV for parameter tunning of the Base algorithm.

```python
from sklearn.model_selection import GridSearchCV
KNN_r = KNeighborsRegressor()
param_grid = {'n_neighbors':[2,4,6,8,10],
              'weights':['uniform', 'distance'],
              'algorithm':['auto', 'ball_tree','kd_tree','brute']}
GCV_KNN_r = GridSearchCV(estimator=KNN_r,param_grid=param_grid,cv=5)
GCV_KNN_r.fit(X_R_train,y_R_train)
```

```python
print('\n Best parameters = ')
print(GCV_KNN_r.best_params_)
```

```python
print('Best Scores = ',GCV_KNN_r.best_score_)
```

```python
#Feeding the best parameters to the Base algorithm.
final_KNN_r = KNeighborsRegressor(algorithm = 'auto', n_neighbors = 4, weights = 'distance')
final_KNN_r.fit(X_R_train,y_R_train)
pred_KNN_r = final_KNN_r.predict(X_R_test)
print('Score of', final_KNN_r, 'is:', final_KNN_r.score(X_R_train, y_R_train))
print('error:')
print('Mean absolute error:',mean_absolute_error(y_R_test,pred_KNN_r))
print('Mean squared error:',mean_squared_error(y_R_test,pred_KNN_r))
print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_R_test,pred_KNN_r)))
print('r2_score:',r2_score(y_R_test,pred_KNN_r))
```

Comparison of the model's performance.

BEFORE :-
1. Score of KNeighborsRegressor() is: 0.9273082495061039

error:
2. Mean absolute error: 0.09508659617033437
3. Mean squared error: 0.018261773078022295
4. Root Mean Squared Error: 0.3083611456885163
5. r2_score: 0.8783847285306946


AFTER PARAMETER TUNNING :-
1. Score of KNeighborsRegressor(n_neighbors=4, weights='distance') is: 1.0

error:
2. Mean absolute error: 0.08525020610724728
3. Mean squared error: 0.015162001753511399

So, the model score increased from 92% to 100% which is quite considerable; also the errors reduced correspondingly.

Let's use cross validation to determine the underfitting/overfitting of model.

```python
#Cross Validation
from sklearn.model_selection import cross_val_score
score_r = cross_val_score(final_KNN_r, X_R_train, y_R_train, cv=5, scoring='r2')
print('Score:',score_r)
print('Mean Score:',score_r.mean())
print('Standard deviation:',score_r.std())
```

Now that the scoring parameter is "r2_score" we have 89% r2_score for our tunned model and after performing cross validation we have mean r2_score of 88%; it clearly states that our model is neither overfitting or underfitting and we have obtained a realistic r2_score.

We can say that now our model is trained and hence can be use for making predictions.

Let's fit our model!

```python
final_KNN_r.fit(X_R,y_R)
final_KNN_r.score(X_R,y_R)
```

Let's check out how our model is actually working.

```python
sns.distplot(pred_KNN_r, label='Prediction')
sns.distplot(y_R_test, label='Actual')
plt.title('Predicted values v/s Actual values')
plt.legend()
plt.show()
```
